<h2>Scrapping wiki pages and cleaning data</h2>

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
print("Imported!")
```

```python
url = "https://es.wikipedia.org/wiki/Anexo:Barrios_de_Bogot%C3%A1"
data = requests.get(url).text
wiki_data = BeautifulSoup(data, 'lxml')
#wiki_data
```

```python
column_names = ['UPZ_Id','UPZ_Name','Borough','Neighborhood']
bogota = pd.DataFrame(columns = column_names)

content = wiki_data.find('div', class_='mw-parser-output')
table = content.table.tbody
upz_id = 0
upz_name = 0
borough = 0
neighborhood = 0

for tr in table.find_all('tr'):
    i = 0
    for td in tr.find_all('td'):
        if i == 0:
            upz_id = td.text.strip('\n')
            i = i + 1
        elif i == 1:
            upz_name = td.text.strip('\n').replace(']','')
            i = i + 1
        elif i == 2: 
            borough = td.text.strip('\n').replace(']','')
            i = i + 1
        elif i == 3: 
            neighborhood = td.text.strip('\n').replace(']','')
    bogota = bogota.append({'UPZ_Id': upz_id,'UPZ_Name' : upz_name,'Borough': borough,'Neighborhood': neighborhood},ignore_index=True)
```

```python
bogota = bogota.replace('\n',' ', regex=True)
bogota.info()
```

```python
bogota['Borough_Id'] = bogota['Borough'].str[0:2]
bogota['Borough'] = bogota['Borough'].str[3:]
bogota = bogota[bogota.UPZ_Id!= 0]
bogota.head()
```

```python
url2 = "https://es.wikipedia.org/wiki/Anexo:Localidades_de_Bogot%C3%A1"
data2 = requests.get(url2).text
wiki_data2 = BeautifulSoup(data2, 'lxml')
#wiki_data2
```

```python
column_names = ['Id_Borough','Borough','PostalCode','Surface','population','Density']
boroughs = pd.DataFrame(columns = column_names)

content = wiki_data2.find('div', class_='mw-parser-output')
table = content.table.tbody
id_borough = 0
borough = 0
postalcode = 0
Surface = 0
population = 0
density = 0

for tr in table.find_all('tr'):
    i = 0
    for td in tr.find_all('td'):
        if i == 0:
            id_borough = td.text.strip('\n')
            i = i + 1
        elif i == 1:
            borough = td.text.strip('\n').replace(']','')
            i = i + 1
        elif i == 2: 
            postalcode = td.text.strip('\n').replace(']','')
            i = i + 1
        elif i == 3: 
            Surface = td.text.strip('\n').replace(']','')
            i = i + 1
        elif i == 4: 
            population = td.text.strip('\n').replace(']','')
            i = i + 1
        elif i == 5: 
            density = td.text.strip('\n').replace(']','')
    boroughs = boroughs.append({'Id_Borough': id_borough,'Borough':borough,'PostalCode': postalcode,'Surface': Surface,'population': population,'Density': density},ignore_index=True)
```

<h2>Adding Latitude and Longitude Co-ordinates to the DataFrame</h2>

```python
import geocoder
```

```python
def get_latlong(UPZ):
    lati_long_coords = None
    while(lati_long_coords is None):
        g = geocoder.arcgis('{}, Bogotá, Colombia'.format(UPZ))
        lati_long_coords = g.latlng
    return lati_long_coords
```

```python
UPZs = bogota['UPZ_Name']    
coords = [ get_latlong(UPZ) for UPZ in UPZs.tolist() ]
df_coords = pd.DataFrame(coords, columns=['Latitude', 'Longitude'])
bogota['Latitude'] = df_coords['Latitude']
bogota['Longitude'] = df_coords['Longitude']
```

<h2>Venue information from Foursquare</h2>

```python
import config as cfg
LIMIT = 70

CLIENT_ID = cfg.credentials['CLIENT_ID'] # your Foursquare ID
CLIENT_SECRET = cfg.credentials['CLIENT_SECRET'] # your Foursquare Secret
VERSION = '20180605' # Foursquare API version
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=700):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
bogota_venues = getNearbyVenues(names=bogota['UPZ_Name'],
                                   latitudes=bogota['Latitude'],
                                   longitudes=bogota['Longitude'],
                                  )
```

<h3>Boroughs data preparation</h3>

```python
boroughs_onehot = pd.get_dummies(borough_venues[['Venue Category']], prefix="", prefix_sep="")

boroughs_onehot['Borough'] = borough_venues['Borough'] 

fixed_columns = [boroughs_onehot.columns[-1]] + list(boroughs_onehot.columns[:-1])
boroughs_onehot = boroughs_onehot[fixed_columns]
boroughs_onehot.shape
```

```python
boroughs_grouped = boroughs_onehot.groupby('Borough').mean().reset_index()
boroughs_grouped
```

```python
num_top_venues = 5
for neigh in boroughs_grouped['Borough']:
    print("----"+neigh+"----")
    temp = boroughs_grouped[boroughs_grouped['Borough'] == neigh].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
import numpy as np
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10
indicators = ['st', 'nd', 'rd']

columns = ['Borough']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

boroughs_venues_sorted = pd.DataFrame(columns=columns)
boroughs_venues_sorted['Borough'] = boroughs_grouped['Borough']

for ind in np.arange(boroughs_grouped.shape[0]):
    boroughs_venues_sorted.iloc[ind, 1:] = return_most_common_venues(boroughs_grouped.iloc[ind, :], num_top_venues)

print(boroughs_venues_sorted.shape)
boroughs_venues_sorted
```
